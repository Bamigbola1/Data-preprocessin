Importing of the Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading of the excel file HR dataset into Jupyter .

```python
data = pd.read_excel ('HRDataset.xlsx')
data
```

Clearing of the missing data in salary column.

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(data.iloc [:, 1:3])
data.iloc[:, 1:3]=imputer.transform(data.iloc[:, 1:3])
data
```

Applying of onehotencoder in Marital Desc column 

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[9])], remainder='passthrough')
data=pd.DataFrame(ct.fit_transform(data))
data
```

Applying of labelencoding on PerformanceScore.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.iloc[:, 86]=le.fit_transform(data.iloc[:, 86])
data
```
